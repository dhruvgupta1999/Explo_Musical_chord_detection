```python
###########################################
#####################This is the final code for the machine learning part of the program
from scipy.stats import rankdata
from numpy.random import seed
seed(3)

# print(test_labels)


for k in range (2,3):
    
    compare_neuron_train_acc=[]
    compare_neuron_val_acc=[]
    print("\n\nvalidation:total ratio",0.1*k)
    for i in range(24,29):  
        print("\nno. of hidden neurons is equal to:",i)
        rank_list=[0]*(no_of_features+1)
        model = Sequential([
            Dense(12, input_shape=train_samples.shape[1:], activation='relu'),
            Dense(i, activation='relu'),
            #Dense(24, activation='relu'),
            layers.Dropout(0.4),
            Dense(i, activation='relu',kernel_regularizer=keras.regularizers.l2(0.06)),
            Dense(12, activation='softmax')
        ])
        
        # load weights
        filepath="weights"+ str(i)+".best.hdf6"
        model.load_weights(filepath)
#         sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=False)
        model.compile(optimizer='adadelta', loss='categorical_crossentropy', metrics=['accuracy'])
        
        # checkpoint
        
        checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=0, save_best_only=True, mode='max')
        callbacks_list = [checkpoint]

        
        history=model.fit(train_samples, train_labels, epochs=100, shuffle=True,callbacks=callbacks_list, verbose=0,validation_split=0.1*k)
#         history=model.fit(train_samples, train_labels, epochs=3000, shuffle=True, verbose=0,validation_split=0.1*k)
        print(history.history.keys())
    
        #  "Accuracy"
        plt.plot(history.history['acc'])
        compare_neuron_train_acc.append(history.history['acc'][-1])
        plt.plot(history.history['val_acc'])
        compare_neuron_val_acc.append(history.history['val_acc'][-1])
        plt.axhline(y=0.9, color='r', linestyle='-')
        plt.axhline(y=0.95, color='r', linestyle='-')
        plt.title('model accuracy at '+str(i)+" nodes per hidden layer")
        plt.ylabel('accuracy')
        plt.xlabel('epoch')
        plt.legend(['train', 'validation'], loc='upper left')
        plt.show()
        # "Loss"
        plt.plot(history.history['loss'])
        plt.plot(history.history['val_loss'])
        plt.title('model loss at '+str(i)+" nodes per hidden layer")
        plt.ylabel('loss')
        plt.xlabel('epoch')
        plt.legend(['train', 'validation'], loc='upper left')
        plt.show()

        prediction=model.predict(test_samples)
#         print(prediction)
        for j in range(prediction.shape[0]):
            tmp= len(prediction[j])-rankdata(prediction[j]).astype(int)+1 #rankdata labels data in increasing order, i.e lowest val=1,highest val=13 in ths case 
#             print("pred[j] is:",prediction[j])
#             print("rankdata[pred[j]] is:",rankdata(prediction[j]).astype(int)+1)
            
#             print("tmp is",tmp)
            tmp=tmp[int(np.nonzero(test_labels[j])[0])]#tmp now contains "rank of prediction" of actual label,i.e test_label[j]
#             print("new_tmp is:",tmp)
            if tmp>2:
                print(test_labels[j])
            rank_list[tmp]+=1 #rank_list is initially list of 13 zeroes,
        for j in range(1,len(rank_list)):
            rank_list[j]+=rank_list[j-1]
        rank_list[:] = [x / test_labels.shape[0] for x in rank_list]
        plt.plot(rank_list)
        plt.axhline(y=0.9, color='r', linestyle='-')
        plt.axhline(y=0.95, color='r', linestyle='-')
        plt.title('CMC curve')
        plt.ylabel('test accuracy at '+str(i)+' nodes per hidden layer')
        plt.xlabel('rank')
        
        plt.show()
```

```python
#code for comparing accuracy among models having different number of nodes
plt.plot(compare_neuron_train_acc)
plt.plot(compare_neuron_val_acc)
plt.axhline(y=0.9, color='r', linestyle='-')
plt.axhline(y=0.95, color='r', linestyle='-')
plt.title('accuracy comparison with changing hidden number of nodes')
plt.ylabel('accuracy')
plt.xlabel('# of hidden nodes')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
